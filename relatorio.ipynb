## Organização dos dados utilizados

Os dados utilizados durante todo o sistema consiste majoritariamente na utilização de listas e dicionários, dando mais coesão ao conceito da colônia utilizada no projeto. A estrutura principal, onde os dados se convergem, sendo fonte de análise geral, é um dicionário, que contém todas as informações utilizadas, como a quantidade de energia reserva, os módulos de geração de energia (solar e eólica), as seções da colônia (como áreas residenciais, medicinais, científicas e logísticas) assim como seus respectivos módulos contendo seus tipos, nível de prioridade, consumo energético e uma flag que diz se o módulo está habilitado ou não.

Após a construção da estrutura da colônia, há também a definição das funções auxiliares, que fazem todos os cálculos, geram dados históricos, pegam os valores energéticos e desabilita os módulos de menor prioridade, conforme necessidade, além da função que mostra o resultado final, que tem o intuito apenas de exibir um relatório final com as informações geradas e a tomada de decisão baseando-se no consumo e produção de energia.

Todas essas estruturas estão separadas em arquivos próprios, sendo as funções utilitárias reservadas no arquivo `utils.py`, exibição do relatório final em `display.py` e o arquivo principal, contendo a estrutura da colônia e utilizando as funções auxiliares sendo o arquivo `main.py`.

## Regras de negócio e tomada de decisão

Foi optado por utilizar regras de negócio simples, para manter a simplicidade do projeto. Pensando nisso, temos como regra:

- Se geração de energia **≥** consumo total: _A colônia tem energia suficiente para a próxima semana_.
  - Se geração de energia **-** consumo total **>** 0: _Armazenar energia excedente_.
- Se geração de energia **<** consumo total: _Atenção: energia produzida é insuficiente para atender ao consumo da próxima semana_.
  - Se energia reserva **≥** consumo total **-** geração de energia: _Reserva de energia utilizada_.
  - Se energia reserva **<** consumo total **-** geração de energia: _Desabilitando módulos de baixa prioridade_.
    - Desabilita todos os módulos de prioridade **baixa** de todas as seções da colônia.
    - Desabilita todos os módulos de prioridade **média** de todas as seções da colônia.
    - Desabilita todos os módulos de prioridade **alta** de todas as seções da colônia.
    - É verificado a cada passo se o consumo energético está inferior à geração de energia e o processo de desabilitação é parado assim que a condição for satisfeita.
    - É mostrado no relatório final quais níveis de prioridade foram desabilitados para manter a colônia funcionando.
    - Módulos de prioridade máxima (0) se mantém ligados independente do consumo pois são tratados como fundamentais (como suporte à vida).

O consumo de energia, ainda que fictício, é representado semanalmente e, portanto, o histórico de energia é gerado utilizando uma medida diária em uma lista com 7 dias.

## Aplicando previsão

No algoritmo, é definido alguns parâmetros para o cálculo da energia gerada utilizando os valores do histórico dos valores solares e eólicos:

- **Constante de energia gerada:** 50;
- **Frequência de geração:**
  - **Solar:** 0.6;
  - **Eólica:** 0.8;
- **Valor histórico:** Inicialmente gerados valores aleatórios entre 0 e 20.

O cálculo da energia gerada é dada por:

$$
  (Valor \ histórico \ \times \ Constante \ de \ energia) \ \times \ Frequência \ de \ geração \ = \ Valor \ energético
$$

Assim, é gerado um valor energético para cada item da lista do valor histórico, ou seja, para cada dia, que somados dá o valor semanal de geração energética, que posteriormente é utilizado para comparação com o valor de consumo semanal da colônia.

Tendo esses valores em mãos, podemos simular uma situação possível. Dado o seguinte valor histórico:

~~~python
history = [2, 4, 6, 8, 10, 12, 14]
~~~

Temos os seguintes valores energéticos utilizando geração de energia solar, aplicando a fórmula acima:

~~~python
energy = [60, 120, 180, 240, 300, 360, 420]
~~~

Sendo assim, fica claro a progressão aritmética do valor histórico ambiental solar (e consequentemente da geração de energia) onde em algum momento atingirá o seu ápice antes de iniciar a queda natural dos valores.

Aplicando a fórmula acima, podemos dizer com precisão quais são os valores de saída baseado nos valores de entrada. Por exemplo, se quisermos prever o dia seguinte ao último valor do histórico energético (14) deduzindo que seguirá o coeficiente 2 da progressão aritmética, temos o valor histórico 16 aplicado no oitavo dia, sendo assim, aplicamos a fórmula da seguinte forma:

$$
  \begin{align}
    &X = (16 \times 50) \times 0.6 \\
    &X = 800 \times 0.6 \\
    &X = 480
  \end{align}
$$

Sendo assim, temos que o valor previsto para um dia que gere 16 no valor histórico gerará 480 de valor energético.

## Assistência algorítmica à eficiência energética da colônia

Entendemos que este é um estágio inicial da colônia em solo marciano, então é esperado que todos os módulos sejam importantes no sistema como um todo, portanto, ainda assim, existe uma categorização por nível de prioridade, que define o quão fundamental é o módulo para o funcionamento da colônia e sobrevivência humana em Marte.

Ao efetuar o cálculo de geração energética, o valor de geração semanal é comparado com o valor de consumo semanal, gerando crédito ou débito energético.

Em caso de excedente, a energia restante é armazenada na reserva da colônia, que tem um valor máximo de 5 mil kWh, dando mais segurança para o funcionamento normal para as próximas semanas.

Caso haja débito, o algoritmo utilizará a reserva da colônia para garantir o funcionamento de todos os módulos normalmente, porém, caso ainda haja débito após o uso da reserva, o algoritmo passa a fazer uma varredura nas seções e desligar os módulos baseando-se no nível de prioridade: primeiro é desligado todos os módulos de prioridade baixa. Caso a colônia ainda esteja com saldo energético negativado, o algoritmo passa a desligar os módulos de nível médio. O mesmo processo ocorre até que a energia gerada mais reserva seja maior do que o consumo, deixando somente os módulos de prioridade máxima sempre ativos, como por exemplo os suportes à vida.

Dessa forma, o algoritmo busca sempre manter saldo positivo na relação entre geração e consumo de energia enquanto alimenta uma reserva sempre que possível.
